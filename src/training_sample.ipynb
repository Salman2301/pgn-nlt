```python
from transformers import pipeline
from datasets import load_dataset
import soundfile as sf
import torch
```

```python

synthesiser = pipeline("text-to-speech", "microsoft/speecht5_tts")

embeddings_dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")
speaker_embedding = torch.tensor(embeddings_dataset[7306]["xvector"]).unsqueeze(0)
```

```python
speech = synthesiser(
"""
a1 b1 c1 d1 e1 f1 g1 h1
a2 b2 c2 d2 e2 f2 g2 h2
a3 b3 c3 d3 e3 f3 g3 h3
a4 b4 c4 d4 e4 f4 g4 h4
a5 b5 c5 d5 e5 f5 g5 h5
a6 b6 c6 d6 e6 f6 g6 h6
a7 b7 c7 d7 e7 f7 g7 h7
a8 b8 c8 d8 e8 f8 g8 h8
pawn king queen bishop knight rook
""", forward_params={"speaker_embeddings": speaker_embedding})
sf.write("../sample/micro-tts.wav", speech["audio"], samplerate=speech["sampling_rate"])
```

```python
import soundfile as sf


synthesiser = pipeline("text-to-speech", "suno/bark-small")
speech = synthesiser("""
a1 b1 c1 d1 e1 f1 g1 h1
a2 b2 c2 d2 e2 f2 g2 h2
a3 b3 c3 d3 e3 f3 g3 h3
a4 b4 c4 d4 e4 f4 g4 h4
a5 b5 c5 d5 e5 f5 g5 h5
a6 b6 c6 d6 e6 f6 g6 h6
a7 b7 c7 d7 e7 f7 g7 h7
a8 b8 c8 d8 e8 f8 g8 h8
pawn king queen bishop knight rook
""", forward_params={"do_sample": True})
sf.write("../sample/suno.wav", speech["audio"], speech["sampling_rate"], subtype='PCM_16')
```

```python
import soundfile as sf
from transformers import pipeline

# Using the gpt2-tts model
synthesiser = pipeline("text-to-speech", model="suno/bark-small")

speech = synthesiser("Hello, my dog is cooler than you!", forward_params={"do_sample": True})

# Write the audio data with 16-bit PCM encoding using the soundfile library
sf.write("../sample/sudo-sf.wav", speech["audio"], speech["sampling_rate"], format='WAV', subtype='PCM_16')
```
