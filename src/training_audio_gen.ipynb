Generate audio based on training.txt file


```python
from gtts import gTTS
from tqdm import tqdm
import os
```

```python
file_dir = "../training"
audio_dir = f"{file_dir}/audio"

# Create the directory if it doesn't exist
if os.path.exists(audio_dir) is not True:
  os.makedirs(audio_dir)
```

```python
file_dir = "../training"
training_path = f"{file_dir}/training.txt"

def create_audio_file(text, file_path, language='en'):
  tts = gTTS(text=text, lang=language, slow=False)
  tts.save(file_path)

  
```

```python
# Get data from the 
training_data = open(training_path, "r").readlines()

for i in tqdm(range(len(training_data))):
  text = training_data[i]
  audio_file_path = f"{audio_dir}/{i}.mp3"
  create_audio_file(text, audio_file_path)
  
print(f"{len_training_data} Audios generated.")
```
